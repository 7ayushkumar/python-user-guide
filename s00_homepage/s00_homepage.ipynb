### What is Plotly?

Answer from Plotly CEO & Co-Founder Jack Parmer:

> Plotly is like graphing crack. It standardizes the graphing interface across scientific computing languages (Python, R, MATLAB, etc) while giving rich interactivity and web shareability that has not been possible before with matplotlib, ggplot, MATLAB, etc. 

> On the Plotly website, you can style your graphs with a GUI, so you don't have to spend hours writing code that simply changes the legend opacity.

> Plotly does this all while backing up your graphs on the cloud, so that years later, you can find data that may have otherwise been on a harddrive in a landfill. If you make your data public, other people can also find your graphs and  data. The best practice that we have today for saving and sharing research data is to entomb it as a thesis in the engineering library basement. All that is changing.

> Like <a href="http://d3js.org/" target="_blank">d3.js</a>? Like interactive, <a href="https://plot.ly/~bchartoff/344/" target="_blank">NYT graphics</a>? So do we. Now, with the Plotly APIs, you can make them yourself without being an expert web programmer. If you are an expert web programmer, now you have scientific languages and tools like  R, Python, Pandas, and MATLAB instead of javascript to wrangle your data and create beautiful data vis. Science meets the world-wide-web. Engineering meets design. Let's do this.

### Why a User Guide?

The User Guide project has the goal of giving Plotly's Python/IPython users extensive documentation on all of Plotly's features. 

Whereas Plotly's web <a href="https://plot.ly/python" target="_blank">documentation</a>  is meant to serve Plotly beginners
just looking to make a simple plot and experienced users trying out a new plot type, 
the User Guide intends to turn beginner and intermediate Plotly users into experts.

It is assumed that readers of the User Guide are *comfortable* with Python terminology (e.g. know what Python objects, lists, dictionaries and functions are). But, there is absolutely no need to be a Python guru to follow the User Guide's content.

### What's new in Plotly's Python API version 1.0?

The User Guide assumes the latest version (v 1.0) of Plotly's Python API verison
1.0. <br> The latest version features several improvements. For instance: 

* A *get* function, to import data and styles from public figures on Plotly's servers, 

* Easier manipulation and customization of Python lists and dictionaries through *graph objects*,

* Running the help function on Plotly objects is now actually helpful,

* Local validation happens on your machine, points directly to errors, and offers suggestions,

* The API now supports local configuration files for saving and retrieving account *credentials*,

* The API now has a subplot generating function.

### Installation guidelines

<h5 style="margin-top:-5px;">Step 1</h5>

The User Guide assumes:

 * Plotly version 1.0.1
 
 * Python version 2.7.5+ with
     - `numpy` installed, for all sections
     - `pandas` installed, for sections 3 and 5
     - `scipy` installed for section 4
 * IPython version 2.0.0 

<h5>Step 2</h5>

To install Plotly, use the package manager 
<a href="https://pypi.python.org/pypi/pip" targer="_blank">pip</a> inside a terminal:

  * &nbsp; &nbsp; &nbsp; &nbsp;`$ pip install plotly`
  * Or &nbsp; `$ sudo pip install plotly`

<h5>Step 3</h5>

Check version (inside Python or IPython), upgrade if needed:

```python
import plotly
print plotly.__version__
```

If not latest version (`1.0.1`), run inside a terminal:

  * `$ pip install plotly --upgrade`

<h5>Step 4</h5>

If you don't already have an account, sign up for Plotly:

  * Go to [plot.ly](https://plot.ly/) and click on the *Sign up* button.

  <img src="http://i.imgur.com/mOknXGB.png" />

<h5>Step 5</h5>

Get username and API key:

  * After signing in, click on the *Settings* tab,
  
  <img src="http://i.imgur.com/6uQmH0L.png" style="margin-top:1em; margin-bottom:1em" />
  
  * Your API key and username are listed under *Profile*.
  
  <img src="http://i.imgur.com/EouTbdP.png" />
  

<h5>Step 6</h5>

For better code portability, we recommend setting up a *credentials* file on your machine:

* In Python or IPython,

 `>>> import plotly.tools as tls` <br>
 `>>> tls.set_credentials_file(username="<->", api_key="<->")`
  
Where the `username` and `api_key` keyword arguments are filled in with your own.

This creates a `.credentials` file in a folder called `$HOME/.plotly/` storing the username and API key locally in JSON file. 

* You can access your credentials in Python/IPython with:

    `>>> my_creds = tls.get_credentials_file()`
    
<h5> Step 7</h5> 

Sign in to Plotly from Python API:

```python
# (New syntax!) tools to communicate with Plotly's server
import plotly.plotly as py  

 # (NEW!) useful Python/Plotly tools 
import plotly.tools as tls 

# Read credentials
my_creds = tls.get_credentials_file() 

# (New syntax!) Sign in to Plotly from API
py.sign_in(my_creds['username'], my_creds['api_key'])
```

*And there you go.* You are now ready to make plots using Plotly and the Python API.

Using the credentials file allows users to share code without having to type in (yet along remember) their own username and API key every time they want to generate a new Plotly plot or modify an existing Plotly plot.
