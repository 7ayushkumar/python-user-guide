Plotly's Streaming API enables your Plotly plots to update in real-time, without refreshing your browser. In other words, users can *continuously* send data to Plotly's servers and visualize this data in *real-time*.

Imagine that you have a thermometer (hooked to an Arduino for example) in your attic and you would like to monitor the temperature readings from your laptop. Plotly together with its Streaming API makes this project easy to achieve.

With Ploly's Streaming API:

> Everyone looking at a Plotly streaming plot sees the same data, updating at the same time. Like all Plotly plots, Plotly streaming plots are immediately shareable, by shortlink or as embedded in website or an IPython notebook. Owners of Plotly plot can edit their with the Plotly web GUI while all of the viewers of the plot will see the changes update live. 

And for the skeptical among us, *it's fast*: graphs update up to 20 times per second.

In this section, we present examples of how to make a Plotly streaming plot.
Readers looking for info about the nuts and bolts of Plotly's streaming API should   refer to <a href="https://plot.ly/streaming/" target="_blank">this link</a>.

### 7.2 Never-ending double pendulum simulation




The Wikipedia <a href="http://en.wikipedia.org/wiki/Double_pendulum" target="_blank">article</a> goes over the equations in details. In essence, we are solving a system of non-linear ODEs (using the `scipy.integrate` module) of the form:

$$ \frac{\partial \boldsymbol{x}}{\partial t} = F(\boldsymbol{x}) \;,$$

where $\boldsymbol{x}$ is the state vector here composed of the angles and angular velocities of both pendulums and 

A system of non-linear ODEs!

The following code has been adapted from the <a href="http://matplotlib.org/examples/animation/double_pendulum_animated.html" target="_blank">Matplotlib example bank</a>. Thanks.

```python
# (*) Import a few functions from numpy, for cleaner code
from numpy import sin, cos, pi

## Define a few constants
G =  9.8      # acceleration due to gravity, in m/s^2
L1 = 1.0      # length of pendulum 1 in m
L2 = 1.0      # length of pendulum 2 in m
M1 = 1.0      # mass of pendulum 1 in kg
M2 = 1.0      # mass of pendulum 2 in kg
rad = pi/180  # 

## Set the initial conditions
th1 = 120.0  # th1 and th2 are the initial angles (degrees)
th2 = -10.0
w1 = 0.0     # w10 and w20 are the initial angular velocities (degrees per second)
w2 = 0.0
state = np.array([th1, w1, th2, w2])*rad  # initial state array

## Time derivatives of each state array variables 
def derivs(state, t):
    dxdt = np.zeros_like(state)   # init. dervative array
    del_ = state[2]-state[0]      # difference of angle1 and angle2
    
    dxdt[0] = state[1]            # derv. of angle1
    
    den1 = (M1+M2)*L1 - M2*L1*cos(del_)*cos(del_)  # deno. of dxdt[2]
    dxdt[1] = (M2*L1*state[1]*state[1]*sin(del_)*cos(del_)
               + M2*G*sin(state[2])*cos(del_) 
               + M2*L2*state[3]*state[3]*sin(del_)
               - (M1+M2)*G*sin(state[0]))/den1    # derv. of angular velo.1

    dxdt[2] = state[3]  # derv. of angle2

    den2 = (L2/L1)*den1  # deno. of dxdt[3]
    dxdt[3] = (-M2*L2*state[3]*state[3]*sin(del_)*cos(del_)  
               + (M1+M2)*G*sin(state[0])*cos(del_)
               - (M1+M2)*L1*state[1]*state[1]*sin(del_)
               - (M1+M2)*G*sin(state[2]))/den2   # derv. of angular velo.2

    return dxdt  # return time derv. array
```

 Now stream data to this plot!
 Write to a plotly stream with plotly-formatted data
 In this case, we'll continually integrate the double pendulum equations 20 points a time, 
 and write the solution, one point at a time

 Our plotly-object is formatted as: {'x': x_data, 'y': y_data}

```python
#
s0 = plotly.stream(stream_tokens[4])

#
s1 = plotly.stream(stream_tokens[5])
```

```python
# (*) 
import scipy.integrate as integrate
import time

## Loop through (infinite loop!)
while True:
    dt = 0.05                 # create a time array from 0..100 
    t = np.arange(0.0,20,dt)  #  sampled at 0.1 second steps
    
    y = integrate.odeint(derivs,state,t)  # solve the system of ODEs!
    
    x1 = L1*sin(y[:,0])       # convert angles to x-y coordinates
    y1 = -L1*cos(y[:,0])
    x2 = L2*sin(y[:,2]) + x1  #   for both pendulums
    y2 = -L2*cos(y[:,2]) + y1
    
    # Write the solutions to Plotly's servers, 1 per stream, 1 point at a time
    for (x1i,y1i,x2i,y2i) in zip(x1,y1,x2,y2):
        s0.write(dict(x=[0, x1i, x2i], y=[0, y1i, y2i]))  # 
        s1.write(dict(x=x2i, y=y2i))                      # 
        time.sleep(0.08)  # plot points 80 ms at a time

    # Set the new initial state
    state = np.array([y[-1,0], y[-1, 1], y[-1, 2], y[-1, 3]])
    
    
```
